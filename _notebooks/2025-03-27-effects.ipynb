## ‚úÖ Beneficial Effects

## ‚ö†Ô∏è Harmful Effects

- **Cyberbullying & Online Risks**  
  Children are vulnerable to scams, inappropriate content, and grooming. Social media also contributes to mental health issues like depression.

- **Social Media Impact**  
  Studies and lawsuits show that apps like TikTok and Instagram can damage teen mental health, cause addiction, and even be fatal due to trends/challenges.

---

## ‚öñÔ∏è Ethical Dilemmas (Neutral Tech)

- **Self-Driving Cars**  
  - *Pros*: Make fast decisions without emotional hesitation.  
  - *Cons*: Can we trust them to make life/death choices?

- **Artificial Intelligence (AI)**  
  - *Pros*: Helps students, boosts productivity, supports healthcare.  
  - *Cons*: Environmental harm, reduced critical thinking, deepfakes.

- **Drones (UAVs)**  
  - *Pros*: Helpful in rescue missions, agriculture, wildfire detection.  
  - *Cons*: Privacy concerns, accident risks, military controversy.

- **Gene Editing**  
  - *Pros*: Could cure genetic diseases and improve agriculture.  
  - *Cons*: Legal/ethical problems, long-term genetic risks.
